```python

def tidyTheRest(df, variable_name):
    columns_list = list(new_sales.columns)
    columns_list[11] = 'OldRevenue'
    columns_list[16] = 'OldQuantity'
    second_char = variable_name[1]
    fifth_char = variable_name[4]
    last_char = variable_name[-1]
    Num = int(second_char) + int(fifth_char) + int(last_char)
    df = df.rename(columns={"Revenue": "OldRevenue"})
    df = df.rename(columns={"Quantity": "OldQuantity"})
    df['Revenue'] = df[columns_list[Num+5]]
    df['Planned revenue'] = df[columns_list[Num+6]]
    df['Product cost'] = df[columns_list[Num+7]]
    df['Quantity'] = df[columns_list[Num+8]]
    df['Unit cost'] = df[columns_list[Num+9]]
    df['Unit price'] = df[columns_list[Num+10]]
    df['Gross profit'] = df[columns_list[Num+11]]
    df['Unit sale price'] = df[columns_list[Num+12]]
    df = df.drop(['Order','method', 'type.1', 'Retailer', 'country', 'OldRevenue', 'Planned', 'revenue', 'Product.3', 'cost', 'OldQuantity', 'Unit', 'cost.1', 'Unit.1', 'price', 'Gross', 'profit', 'Unit.2', 'sale', 'price.1'],axis=1)
    return df

def setRetailerCountry(up_df, df, variable_name):
    columns_list = list(new_sales.columns)
    second_char = variable_name[1]
    fifth_char = variable_name[4]
    last_char = variable_name[-1]
    Num = int(second_char) + int(fifth_char) + 6
    if last_char == '1':
        df = up_df[~up_df[columns_list[Num]].apply(lambda x: isinstance(x, str))].copy(deep=True)
        df['Retailer country'] = df[columns_list[Num-1]]
    if last_char == '2':
        df = up_df[up_df[columns_list[Num]].apply(lambda x: isinstance(x, str))].copy(deep=True)
        df['Retailer country'] = df[columns_list[Num-1]] + ' ' + df[columns_list[Num]]
    return df

p2 = temp_sales[temp_sales['method'].isin([
    'E-mail','Fax','Mail','Sales','Special','Telephone','Web'
])].copy(deep=True)
p2.groupby(['Product.2','Order','method']).size()
p2['Product'] = p2['Product.2'] + ' ' + p2['Order']
p2 = p2.drop(['Product.2'],axis=1)

p2_o1 = p2[p2['method'].isin([
    'E-mail','Fax','Mail','Special','Telephone','Web'
])].copy(deep=True)
p2_o1['Order method type'] =  p2_o1['method']
p2_o1_r1 = p2_o1[~p2_o1['Retailer'].apply(lambda x: isinstance(x, str))].copy(deep=True)
p2_o1_r1['Retailer country'] = p2_o1_r1['type.1']
p2_o1_r1 = tidyTheRest(p2_o1_r1,'p2_o1_r1')
#p2_o1_r2 = p2_o1[p2_o1['Retailer'].apply(lambda x: isinstance(x, str))].copy(deep=True)
#p2_o1_r2['Retailer country'] = p2_o1_r2['type.1'] + ' ' + p2_o1_r2['Retailer']
#p2_o1_r2 = tidyTheRest(p2_o1_r2,'p2_o1_r2')
p2_o1_r2 = pd.DataFrame()
p2_o1_r2 = setRetailerCountry(p2_o1, p2_o1_r2, 'p2_o1_r2')
p2_o1_r2 = tidyTheRest(p2_o1_r2,'p2_o1_r2')


#p2.groupby(['Product.2','Order','method','type.1']).size()
p2_o1_r2.head()
```
