```python
import pandas as pd
import re
```

```python
#第1題
sales = pd.read_excel('./sales.xlsx',sheet_name='Sheet1')
```

```python
#第2題
print(sales.dtypes)
```

```python
#第3題
column_names = sales.columns.tolist()
pattern = re.compile(r'[.\d]+')
for i in range(len(column_names)):
    column_names[i] = re.sub(pattern, '', column_names[i])

new_columns = []
prev_col = ""

for col in column_names:
    if col[0].islower():
        prev_col += " " + col
    else:
        new_columns.append(prev_col)
        prev_col = col

new_columns.append(prev_col)
del new_columns[0]
for col in new_columns:
    print(col)
```

```python
#第4題 「沒錯置到下一欄位，就不會錯置到更後面的欄位」所以Year無錯置問題
sales.groupby(['Year','Product']).size()
```

```python
#第5題

#先檢查錯置情況
print(sales.groupby(['Product','line','Product.1']).size())

#整理
sales['Product line'] = sales['Product'] + ' ' + sales['line']
sales = sales.drop(['Product','line'], axis=1)
sales.head()
```

```python
#第6題+第7題
sales['Product type'] = sales['Product.1'] + ' ' + sales['type']
sales = sales.drop(['Product.1','type'], axis=1)
condition= (sales['Product type']=='Climbing Accessories') | (sales['Product type']=='Cooking Gear') | (sales['Product type']=='First Aid') | (sales['Product type']=='Golf Accessories') | (sales['Product type']=='Insect Repellents') | (sales['Product type']=='Sleeping Bags')
new_sales=sales[condition]

print(new_sales.groupby(['Product type','Product.2']).size())
new_sales.head()
```

```python
#第8題~第11題 - 檢查錯置情況部份


#第8題:持續檢查直到groupby返回值為空，以下發現Produt的長度為1~6沒有到7(p1~p6的檢查之程式碼在此處刪除而移到後方各自區塊)
p7 = new_sales[new_sales['Planned'].isin([
    'E-mail','Fax','Mail','Sales','Special','Telephone','Web'
])].copy(deep=True)
print('p7:')
print(p7.groupby(['Product.2','Order','method','type.1','Retailer','country','Revenue','Planned']).size())

#第9題:由題目已知Order method type有7種且長度有1~2

#第10題:後方各區有以檢查後一格資料是否為數字判斷出Retailer country長度(均為1~2)

#第11題:剩下資料皆為數字所占長度均為1
```

```python
#第8題~第11題
#本區塊將重複性高、繁冗的部分寫成function(第9題、第10題)
ori_sales = pd.read_excel('./sales.xlsx',sheet_name='Sheet1')

#此function負責整理Retailer country(第9題)
def setRetailerCountry(up_df, df, variable_name):
    columns_list = list(ori_sales.columns)
    second_char = variable_name[1]
    fifth_char = variable_name[4]
    last_char = variable_name[-1]
    Num = int(second_char) + int(fifth_char) + 6
    if last_char == '1':
        df = up_df[~up_df[columns_list[Num]].apply(lambda x: isinstance(x, str))].copy(deep=True)
        df['Retailer country'] = df[columns_list[Num-1]]
    if last_char == '2':
        df = up_df[up_df[columns_list[Num]].apply(lambda x: isinstance(x, str))].copy(deep=True)
        df['Retailer country'] = df[columns_list[Num-1]] + ' ' + df[columns_list[Num]]
    return df

#此function負責整理剩下之數字欄位(第10題)
def tidyTheRest(df, variable_name):
    columns_list = list(ori_sales.columns)
    columns_list[11] = 'OldRevenue'
    columns_list[16] = 'OldQuantity'
    second_char = variable_name[1]
    fifth_char = variable_name[4]
    last_char = variable_name[-1]
    Num = int(second_char) + int(fifth_char) + int(last_char)
    df = df.rename(columns={"Revenue": "OldRevenue"})
    df = df.rename(columns={"Quantity": "OldQuantity"})
    df['Revenue'] = df[columns_list[Num+5]]
    df['Planned revenue'] = df[columns_list[Num+6]]
    df['Product cost'] = df[columns_list[Num+7]]
    df['Quantity'] = df[columns_list[Num+8]]
    df['Unit cost'] = df[columns_list[Num+9]]
    df['Unit price'] = df[columns_list[Num+10]]
    df['Gross profit'] = df[columns_list[Num+11]]
    df['Unit sale price'] = df[columns_list[Num+12]]
    df = df.drop(['Order','method', 'type.1', 'Retailer', 'country', 'OldRevenue', 'Planned', 'revenue', 'Product.3', 'cost', 'OldQuantity', 'Unit', 'cost.1', 'Unit.1', 'price', 'Gross', 'profit', 'Unit.2', 'sale', 'price.1'],axis=1)
    return df
```

```python
#第8~11題
#Product長度為1的整理區(第8題)

p1 = new_sales[new_sales['Order'].isin([
    'E-mail','Fax','Mail','Sales','Special','Telephone','Web'
])].copy(deep=True)
print(p1.groupby(['Product.2','Order']).size())
p1['Product'] = p1['Product.2']
p1 = p1.drop(['Product.2'],axis=1)

#Order method type長度為1的整理區(第9題)
p1_o1 = p1[p1['Order'].isin([
    'E-mail','Fax','Mail','Special','Telephone','Web'
])].copy(deep=True)
p1_o1['Order method type'] =  p1_o1['Order']
p1_o1_r1 = pd.DataFrame()
p1_o1_r1 = setRetailerCountry(p1_o1, p1_o1_r1, 'p1_o1_r1')#Retail country長度為1的整理區(第10題)
p1_o1_r1 = tidyTheRest(p1_o1_r1,'p1_o1_r1')#剩下欄位整理區(第11題)
p1_o1_r2 = pd.DataFrame()
p1_o1_r2 = setRetailerCountry(p1_o1, p1_o1_r2, 'p1_o1_r2')#Retail country長度為2的整理區(第10題)
p1_o1_r2 = tidyTheRest(p1_o1_r2,'p1_o1_r2')

#Order method type長度為2的整理區(第9題)
p1_o2 = p1[p1['Order'].isin([
    'Sales'
])].copy(deep=True)
p1_o2['Order method type'] =  p1_o2['Order'] + ' ' + p1_o2['method']
p1_o2_r1 = pd.DataFrame()
p1_o2_r1 = setRetailerCountry(p1_o2, p1_o2_r1, 'p1_o2_r1')
p1_o2_r1 = tidyTheRest(p1_o2_r1,'p1_o2_r1')
p1_o2_r2 = pd.DataFrame()
p1_o2_r2 = setRetailerCountry(p1_o2, p1_o2_r2, 'p1_o2_r2')
p1_o2_r2 = tidyTheRest(p1_o2_r2,'p1_o2_r2')

p1_o1 = pd.concat([p1_o1_r1,p1_o1_r2])
p1_o2 = pd.concat([p1_o2_r1,p1_o2_r2])
p1 = pd.concat([p1_o1,p1_o2])

p1.head()
###################################
# p2~p6區與p1區雷同故不過度詳細註解 #
##################################
```

```python
#第8~11題
#Product長度為2的整理區(詳細註解於p1區)

p2 = new_sales[new_sales['method'].isin([
    'E-mail','Fax','Mail','Sales','Special','Telephone','Web'
])].copy(deep=True)
print(p2.groupby(['Product.2','Order','method']).size())
p2['Product'] = p2['Product.2'] + ' ' + p2['Order']
p2 = p2.drop(['Product.2'],axis=1)

p2_o1 = p2[p2['method'].isin([
    'E-mail','Fax','Mail','Special','Telephone','Web'
])].copy(deep=True)
p2_o1['Order method type'] =  p2_o1['method']
p2_o1_r1 = pd.DataFrame()
p2_o1_r1 = setRetailerCountry(p2_o1, p2_o1_r1, 'p2_o1_r1')
p2_o1_r1 = tidyTheRest(p2_o1_r1,'p2_o1_r1')
p2_o1_r2 = pd.DataFrame()
p2_o1_r2 = setRetailerCountry(p2_o1, p2_o1_r2, 'p2_o1_r2')
p2_o1_r2 = tidyTheRest(p2_o1_r2,'p2_o1_r2')

p2_o2 = p2[p2['method'].isin([
    'Sales'
])].copy(deep=True)
p2_o2['Order method type'] =  p2_o2['method'] + ' ' + p2_o2['type.1']
p2_o2_r1 = pd.DataFrame()
p2_o2_r1 = setRetailerCountry(p2_o2, p2_o2_r1, 'p2_o2_r1')
p2_o2_r1 = tidyTheRest(p2_o2_r1,'p2_o2_r1')
p2_o2_r2 = pd.DataFrame()
p2_o2_r2 = setRetailerCountry(p2_o2, p2_o2_r2, 'p2_o2_r2')
p2_o2_r2 = tidyTheRest(p2_o2_r2,'p2_o2_r2')

p2_o1 = pd.concat([p2_o1_r1,p2_o1_r2])
p2_o2 = pd.concat([p2_o2_r1,p2_o2_r2])
p2 = pd.concat([p2_o1,p2_o2])

p2.head()
```

```python
#第8~11題
#Product長度為3的整理區(詳細註解於p1區)

p3 = new_sales[new_sales['type.1'].isin([
    'E-mail','Fax','Mail','Sales','Special','Telephone','Web'
])].copy(deep=True)
print(p3.groupby(['Product.2','Order','method','type.1']).size())
p3['Product'] = p3['Product.2'] + ' ' + p3['Order'] + ' ' + p3['method']
p3 = p3.drop(['Product.2'],axis=1)

p3_o1 = p3[p3['type.1'].isin([
    'E-mail','Fax','Mail','Special','Telephone','Web'
])].copy(deep=True)
p3_o1['Order method type'] =  p3_o1['type.1']
p3_o1_r1 = pd.DataFrame()
p3_o1_r1 = setRetailerCountry(p3_o1, p3_o1_r1, 'p3_o1_r1')
p3_o1_r1 = tidyTheRest(p3_o1_r1,'p3_o1_r1')
p3_o1_r2 = pd.DataFrame()
p3_o1_r2 = setRetailerCountry(p3_o1, p3_o1_r2, 'p3_o1_r2')
p3_o1_r2 = tidyTheRest(p3_o1_r2,'p3_o1_r2')

p3_o2 = p3[p3['type.1'].isin([
    'Sales'
])].copy(deep=True)
p3_o2['Order method type'] =  p3_o2['type.1'] + ' ' + p3_o2['Retailer']
p3_o2_r1 = pd.DataFrame()
p3_o2_r1 = setRetailerCountry(p3_o2, p3_o2_r1, 'p3_o2_r1')
p3_o2_r1 = tidyTheRest(p3_o2_r1,'p3_o2_r1')
p3_o2_r2 = pd.DataFrame()
p3_o2_r2 = setRetailerCountry(p3_o2, p3_o2_r2, 'p3_o2_r2')
p3_o2_r2 = tidyTheRest(p3_o2_r2,'p3_o2_r2')

p3_o1 = pd.concat([p3_o1_r1,p3_o1_r2])
p3_o2 = pd.concat([p3_o2_r1,p3_o2_r2])
p3 = pd.concat([p3_o1,p3_o2])

p3.head()
```

```python
#第8~11題
#Product長度為4的整理區(詳細註解於p1區)

p4 = new_sales[new_sales['Retailer'].isin([
    'E-mail','Fax','Mail','Sales','Special','Telephone','Web'
])].copy(deep=True)
print(p4.groupby(['Product.2','Order','method','type.1','Retailer']).size())
p4['Product'] = p4['Product.2'] + ' ' + p4['Order'] + ' ' + p4['method'] + ' ' + p4['type.1']
p4 = p4.drop(['Product.2'],axis=1)

p4_o1 = p4[p4['Retailer'].isin([
    'E-mail','Fax','Mail','Special','Telephone','Web'
])].copy(deep=True)
p4_o1['Order method type'] =  p4_o1['Retailer']
p4_o1_r1 = pd.DataFrame()
p4_o1_r1 = setRetailerCountry(p4_o1, p4_o1_r1, 'p4_o1_r1')
p4_o1_r1 = tidyTheRest(p4_o1_r1,'p4_o1_r1')
p4_o1_r2 = pd.DataFrame()
p4_o1_r2 = setRetailerCountry(p4_o1, p4_o1_r2, 'p4_o1_r2')
p4_o1_r2 = tidyTheRest(p4_o1_r2,'p4_o1_r2')

p4_o2 = p4[p4['Retailer'].isin([
    'Sales'
])].copy(deep=True)
p4_o2['Order method type'] =  p4_o2['Retailer'] + ' ' + p4_o2['country']
p4_o2_r1 = pd.DataFrame()
p4_o2_r1 = setRetailerCountry(p4_o2, p4_o2_r1, 'p4_o2_r1')
p4_o2_r1 = tidyTheRest(p4_o2_r1,'p4_o2_r1')
p4_o2_r2 = pd.DataFrame()
p4_o2_r2 = setRetailerCountry(p4_o2, p4_o2_r2, 'p4_o2_r2')
p4_o2_r2 = tidyTheRest(p4_o2_r2,'p4_o2_r2')

p4_o1 = pd.concat([p4_o1_r1,p4_o1_r2])
p4_o2 = pd.concat([p4_o2_r1,p4_o2_r2])
p4 = pd.concat([p4_o1,p4_o2])

p4.head()
```
