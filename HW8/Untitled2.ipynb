```python
import pickle
import pandas as pd

# 導入CSV檔案並讀取資料
with open('Apk.csv', 'rb') as file:
    data = pickle.load(file)

# 將資料轉換為DataFrame
df = pd.DataFrame(data)
df = df.drop('apk', axis=1)
df.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('Permission')),columns=mlb.classes_,index=df.index))
df= df.replace({'benign': 0, 'Banking': 1,'Adware':1,'SMS':1})
df.head()
```

```python
df.to_csv('output_3.csv', index=False)
```
