```python
import pickle
import pandas as pd

# 導入CSV檔案並讀取資料
with open('Apk.csv', 'rb') as file:
    data = pickle.load(file)

# 將資料轉換為DataFrame
df = pd.DataFrame(data)
df = df.drop('apk', axis=1)
df.head()
```

```python
# 創建空的權限列表
all_permissions = []

# 遍歷每筆資料的Permission列表
for permissions in df['Permission']:
    for permission in permissions:
        if permission not in all_permissions:
            all_permissions.append(permission)

# 創建新的權限特徵列
for permission in all_permissions:
    df[permission] = df['Permission'].apply(lambda x: 1 if permission in x else 0)

# 刪除原始Permission列
df = df.drop('Permission', axis=1)

# 查看處理後的DataFrame
df.head()
```

```python
# 將Class屬性轉換為0和1
df['Class'] = df['Class'].apply(lambda x: 0 if x == 'benign' else 1)

# 查看處理後的DataFrame
df.head()
```

```python
# 將所有屬性轉換為名義型
df = df.astype(object)

# 查看處理後的DataFrame
df.head()
```

```python
from PIL import Image
import numpy as np

# 假設data是包含所有數據的二維數組，每一行代表一筆數據
# data = [...]

# 獲取數據的形狀
num_data_points, num_features = df.shape

# 假設數據是正方形的圖片，則寬度和高度相等
width = height = int(np.sqrt(num_features))

# 將數據轉換為二進制矩陣
binary_matrix = np.array(df)

# 轉換矩陣形狀
reshaped_matrix = binary_matrix.reshape((-1, width, height))

# 將矩陣轉換為圖片
image = Image.fromarray(reshaped_matrix[0], mode='L')

# 顯示圖片
image.show()
```
