```python
import pickle
import pandas as pd

# 導入CSV檔案並讀取資料
with open('Apk.csv', 'rb') as file:
    data = pickle.load(file)

# 將資料轉換為DataFrame
df = pd.DataFrame(data)
df = df.drop('apk', axis=1)
df.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

# 假设您的数据框名为df，其中包含Class和Permission列

# 创建MultiLabelBinarizer对象
mlb = MultiLabelBinarizer()

# 将Permission列转换为多标签编码
encoded_permissions = mlb.fit_transform(df['Permission'])

# 将编码后的权限转换为数据框
encoded_permissions_df = pd.DataFrame(encoded_permissions, columns=mlb.classes_)

# 将编码后的权限与原始数据框合并
df_encoded = pd.concat([df['Class'], encoded_permissions_df], axis=1)

# 打印转换后的数据框
df_encoded.head()
```

```python
# 將Class屬性轉換為0和1
df_encoded['Class'] = df_encoded['Class'].apply(lambda x: 0 if x == 'benign' else 1)

# 查看處理後的DataFrame
df_encoded.head()
```

```python
df_encoded.to_csv('output_encoded.csv', index=False)
```
